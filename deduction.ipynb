```python
from nltk.inference import ResolutionProver
from nltk.sem import Expression

# Facts and rule:
# 1. All humans are mortal.
# 2. Socrates is a human.
# Deduce: Socrates is mortal.

# Parse the logical expressions
read_expr = Expression.fromstring
goal = read_expr('mortal(socrates)')
clauses = [read_expr('human(socrates)'), read_expr('all x (human(x) -> mortal(x))')]

print(ResolutionProver().prove(goal, clauses))
```
